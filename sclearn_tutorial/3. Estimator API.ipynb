# Estimator API (application programming interface)

- It provides a consistent interface for a wide range of ML applications.
- All ML algorithms in sklearn are implemented via Estimator API
- The object that learns from the data is an **estimator**
---


- **estimator.fit(data)** for fitting data
- **Estimator(param1 = 1, param2 = 2)** all the paras of the estimator can be set
- **estimator.estimated_param_** once the data is fitted with an estimator, we can access the estimated paras

### Estimation and Decoding of a model

The model is estimated by:
- paras which are provided in object construction
- the global random state (np.random) if the random_state = none
- data passed to **fit(), fit_transform(), fit_predict()**
- data passed in a sequence of calls to **partial_fit**

### Mapping non-rectangular data representation into rectangular data

### Distinction between core and outlying samples

by using:
    - fit()
    - fit_predict() if transductive
    - predict if inductive


# Design a Scikit-learn API

- All the objects should share a common interface drawn from a limited set of methods
- Algorithm should be represented by python classes
- Datasets should be represented in standard format (Numpy Array, pandas dataframes, scipy sparse matrix)

# Using Estimator API

### 1. Choose a class of model
### 2. Set model paras
### 3. Arrange the data
### 4. Model fitting **fit()**
### 5. Applying the model


- Supervised Learning -- predict()
- Unsupervised Learning -- predict() or transform()


```python


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

samples = np.random.RandomState(35)
x = 10 * samples.rand(40)
y = 2 * x - 1 + samples.randn(40)
plt.scatter(x, y)
```

```python
# choose model
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept = True)

X = x.reshape(-1, 1) # [40, 1]
model.fit(X, y)
```

```python
print(f"Slope: {model.coef_}, Intercept: {model.intercept_}")
```

```python
# predict

xfit = np.linspace(-1, 11)
X_test = xfit[:, np.newaxis] # add new axis to make [50, 1]
y_hat = model.predict(X_test)
plt.plot(xfit, y_hat)
plt.scatter(x, y)
```

### Unsupervised Learning Example


```python
import seaborn as sns
from sklearn.decomposition import PCA

iris = sns.load_dataset("iris")
X_iris = iris.drop("species", axis = 1)
y_iris = iris["species"]

model = PCA(n_components = 2)
model.fit(X_iris)
X_2D = model.transform(X_iris)

species = iris['species'].unique()

for sp in species:
    plt.scatter(X_2D[y_iris == sp, 0], X_2D[y_iris == sp, 1], label=sp)

plt.legend()
plt.show()
```
