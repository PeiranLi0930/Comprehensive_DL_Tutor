```python
import sklearn
import sklearn.datasets as ds
import seaborn

iris = seaborn.load_dataset("iris")
iris.head()
```

# Visualizing the Data

### Distribution of Individual Features (Diagonal Plots):

- **Histograms**: Show the frequency distribution of each feature. For example, you can see that the sepal length has a roughly normal distribution, while the petal length and petal width show a bimodal distribution, indicating two groups within the data.
### Relationships Between Features (Off-Diagonal Plots):

- **Scatter Plots**: Each plot shows the relationship between two features. If the points form a distinct pattern, like a line, this suggests a correlation between the variables. For example, petal length and petal width display a strong positive correlation, as the points suggest a linear relationship.

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
seaborn.set()
seaborn.pairplot(iris, hue = "species", height = 3)
plt.show()
```
