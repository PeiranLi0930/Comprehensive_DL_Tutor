# Introduction to [Matplotlib](https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
plt.plot(range(10))
plt.show()
```

`plt.show()` function preoccupies the interpreter with the controls in the GUI and blocks further interaction. 
- Ipython provides ways to get around this blocking, so we can simultaneously interact with the interpreter and the figure window
    - Can do this in the plain Python interpreter by doing `import matplotlib; matplotlib.interactive(True)`

`plt.show()` provokes the embedded artists to render on the Matplotlib canvas.

## Alternatives to Matplotlib
- **PyQtGraph**: real-time data display and tools for volumetric data rendering and complicated 3D meshes with isosurfaces.
- **ggplot**: tight integration with the Pandas Dataframe, which makes it easy to draw beautifully formatted statistical graphs.

## Extensions to Matplotlib

`seaborn` module includes a vast array of beautifully formatted plots, whose gallery includes samples of available plots and the corresponding code that generates them
- Importing `seaborn` hijacks the default setting for all plots, so you have to coordinate this if you only want to use it for some of the visualizations in a given session.
    - The defaults for Matplotlib is in **`matplotlib.rcParams`** dictionary.
